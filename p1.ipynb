```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def analytical(k, b, t):
    C1 = b
    C2 = -b/k
    return b*t/k + C1 * np.exp(-t)/k + C2

def numerical(k, b, t, y_prev, y_curr, n = 1, y_next = 0):
    if len(t) - 1 == n:
        return y_next
    y_next = (b * t**2 + (t * k + 2) * y_curr - y_prev) / (t * k + 1)
    y_prev = y_curr
    y_curr = y_next
    return numerical(k, b, t, y_curr, y_next, n + 1, y_next)
```

```python
x = np.arange(0, 10, 0.01)
k = 10**(-3)
b = 10
#plt.plot(x, analytical(k, b, x), label = "Analytical")
plt.plot(x, numerical(k, b, x, 0, 0), label="Numerical", linestyle='--')
plt.legend()
plt.xlabel("t")
plt.ylabel("y(t)")
# plt.xlim([0, 1])
# plt.ylim([-10, 10])
plt.grid(True)
```
